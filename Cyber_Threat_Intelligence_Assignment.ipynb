```python
import requests

# Fecthing data from url
url = 'https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json'
resonse = requests.get(url)
data = resonse.json()
```

```python
# Task 1: Extracting the objects where type == 'attack-pattern' and storing them into a list called techniques
techniques = []
for record in data['objects']:
  if record['type'] == 'attack-pattern':
    techniques.append(record)
print(len(techniques))
```

```python
# Task 2a : Threat Scoring Logic
# Evaluting threat scores based on two factors Base Score and Name-Based Intelligence
def threat_score(techniques):
    for tech in techniques:
        score = 5   # Base score for everyone

        name = tech['name'].lower()

        if 'credential' in name:
            score += 3
        elif 'execution' in name:
            score += 2
        elif 'privilege' in name:
            score += 2
        elif 'persistence' in name:
            score += 2
        elif 'lateral' in name:
            score += 2

        # Always assign the score
        tech['score'] = score

    return techniques
result = threat_score(techniques)
```

```python
# Task 2b: Description-Based Intelligence
# Analysing the descrption field of each technique and increased the score if it has matching keywords
def descriptionBased_score(techniques):
  for tech in techniques:
    description = tech['description'].lower()

    if 'administrator' in description:
      tech['score'] += 2

    if 'remote' in description:
      tech['score'] += 2

    if 'bypass' in description:
      tech['score'] += 2

    if 'stealth' in description:
      tech['score'] += 1

  return techniques

result1 = descriptionBased_score(techniques)
```

```python
# Task 3: store techniques in a list called scoredTechniques
def scored_techniques(techniques):
  scoredTechniques = []
  for tech in techniques:
    scoredTechniques.append([tech['name'], tech['score']])
  return scoredTechniques

scoredTechniques =  scored_techniques(techniques)
```

```python
# Task 4: Sort techniques from highest to lowest
# Display top 10 risk techniques
def displayHighestRisk(scoredTechniques):
  scoredTechniques.sort(key=lambda x: x[1], reverse=True)
  for tech in scoredTechniques[:10]:
    print(tech)

displayHighestRisk(scoredTechniques)
```

```python
# Task 5: display techniques greater than 8.9 score
def criticalThreats(techniques):
  for tech in techniques:
    # print(tech['name'], tech['score'])
    if tech['score'] >= 8.5:
      print(tech['name'], tech['score'])

criticalThreats(techniques)
```
